```python
# Creating a Model for Logistic Regression
y = df.target.values
x_data = df.drop(['target'], axis=1)
```

```python
# Normalize the data !Split our data for training and testing
# 80% training # 20% testing
# Normalize the data x =  (X-Xmin) / (Xmax - Xmin)
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
```

```python
# Livingstone
"""
Linear regression
polynomial regression
multiple regression
scale

Jeff
Train/test
logistic regression
decision tree
confusion matrix
hierarchical clustering
categorical data
k-means
cross-validation
AUC-ROC curve
KNN

Independent variables and dependent variables
"""
```

```python
# Split the data 80% training # 20% testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
# Logistic Regresion, testing and training, independent and dependent variables, decistion tree
```
